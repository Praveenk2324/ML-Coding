## 1. Machine Learning Fundamentals

---

### Q1. What is the Bias-Variance Tradeoff?

**Answer:**
The bias-variance tradeoff describes the tension between two sources of error that prevent supervised learning models from generalizing well.

- **Bias**: Error from overly simplistic assumptions in the learning algorithm. High bias → underfitting (model misses relevant relationships in data).
- **Variance**: Error from sensitivity to small fluctuations in the training set. High variance → overfitting (model fits noise, not signal).

The total expected test error can be decomposed as:

$$\text{Total Error} = \text{Bias}^2 + \text{Variance} + \text{Irreducible Noise}$$

| Model | Bias | Variance | Symptom |
|:---|:---|:---|:---|
| Too simple (e.g. Linear on non-linear data) | High | Low | Underfitting |
| Too complex (e.g. Deep tree on small dataset) | Low | High | Overfitting |
| Optimal | Balanced | Balanced | Generalizes well |

**Key Insight**: Increasing model complexity generally decreases bias but increases variance. Regularization, cross-validation, and ensembles help manage this tradeoff.

---

### Q2. What is the difference between parametric and non-parametric models?

**Answer:**

| Property | Parametric | Non-Parametric |
|:---|:---|:---|
| **Assumption** | Assumes a fixed functional form | No fixed form assumed |
| **Parameters** | Fixed number (learned during training) | Grows with data |
| **Examples** | Linear Regression, Logistic Regression, Naive Bayes, Neural Nets | k-NN, Decision Trees, SVMs (with RBF kernel), Gaussian Processes |
| **Pros** | Faster, interpretable, less data needed | Flexible, powerful |
| **Cons** | Bias if form is wrong | Slow on large data, prone to overfitting |

**Intuition**: A parametric model is like fitting a pre-defined curve shape; non-parametric models let the data "speak for itself."

---

### Q3. Explain overfitting and how to prevent it.

**Answer:**
**Overfitting** occurs when a model learns the training data too well — including its noise — and fails to generalize to unseen data. Symptoms: very low training error, high validation/test error.

**Prevention Strategies:**

| Technique | How It Helps |
|:---|:---|
| **More training data** | Reduces variance by exposing model to more patterns |
| **Regularization (L1/L2)** | Penalizes large weights, reducing model complexity |
| **Dropout** (Neural Nets) | Randomly disables neurons during training, forcing redundancy |
| **Early Stopping** | Halts training when validation loss starts rising |
| **Cross-Validation** | Ensures evaluation on multiple unseen splits |
| **Pruning** (Trees) | Reduces tree depth/leaves |
| **Ensemble Methods** | Averaging over models reduces variance (Bagging/Random Forest) |
| **Data Augmentation** | Artificially expands training data |

---

### Q4. What is Cross-Validation and why is it important?

**Answer:**
Cross-validation is a technique to estimate a model's generalization performance on unseen data by evaluating it on multiple held-out partitions of the training data.

**k-Fold Cross-Validation (most common):**
1. Split data into $k$ equally-sized folds.
2. For each fold $i$: train on all folds except $i$, evaluate on fold $i$.
3. Report mean ± std of the $k$ scores.

**Why it matters:**
- Provides a more reliable estimate of test performance than a single train/val split.
- Detects high variance in model performance.
- Enables hyperparameter tuning without "leaking" test data.

**Variants**: Stratified k-Fold (preserves class distribution), Leave-One-Out (LOOCV), Time Series Split.

### Q5. What is the difference between L1 and L2 regularization?

**Answer:**
Both add a penalty term to the loss function to discourage large weights and reduce overfitting.

**L2 Regularization (Ridge)**:
$$J = L_{\text{base}} + \frac{\lambda}{2} \sum_j w_j^2$$
- Penalizes squared magnitude of weights.
- Produces **small but non-zero** weights → smooth solutions.
- Closed-form solution exists.
- Preferred when most features are relevant.

**L1 Regularization (Lasso)**:
$$J = L_{\text{base}} + \lambda \sum_j |w_j|$$
- Penalizes absolute magnitude of weights.
- **Induces sparsity** → drives irrelevant feature weights to exactly zero (built-in feature selection).
- No closed-form; requires subgradient methods.
- Preferred when you believe many features are irrelevant.

**Elastic Net** combines both: $J = L_{\text{base}} + \lambda_1 \|w\|_1 + \lambda_2 \|w\|_2^2$


## 2. Model Evaluation & Metrics

---

### Q6. Explain Precision, Recall, F1-Score and when to use each.

**Answer:**
These are classification metrics derived from the confusion matrix.

Given: **TP** (True Positives), **FP** (False Positives), **FN** (False Negatives), **TN** (True Negatives):

$$\text{Precision} = \frac{TP}{TP + FP}$$
$$\text{Recall (Sensitivity)} = \frac{TP}{TP + FN}$$
$$\text{F1-Score} = 2 \cdot \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$$

| Metric | Prioritize When... | Example Use Case |
|:---|:---|:---|
| **Precision** | Cost of false positives is high | Spam detection (don't flag real email as spam) |
| **Recall** | Cost of false negatives is high | Cancer screening (don't miss a cancer patient) |
| **F1-Score** | You need balance of both | Fraud detection, NER tagging |
| **Accuracy** | Classes are balanced | MNIST digit classification |

**Key trap**: Accuracy is misleading on **imbalanced datasets** (e.g., 99% accuracy on a dataset where 99% of samples are class 0 is meaningless).

---

### Q7. What is AUC-ROC and what does it measure?

**Answer:**
**ROC Curve (Receiver Operating Characteristic)** plots **True Positive Rate (Recall)** vs **False Positive Rate** at various classification thresholds.

$$\text{FPR} = \frac{FP}{FP + TN}, \quad \text{TPR} = \frac{TP}{TP + FN}$$

**AUC (Area Under the ROC Curve)**: A single scalar summarizing the ROC curve, ranging from 0 to 1.

| AUC Value | Interpretation |
|:---|:---|
| 1.0 | Perfect classifier |
| 0.5 | Random classifier (diagonal line) |
| < 0.5 | Worse than random |

**Intuition**: AUC represents the probability that a randomly chosen positive example is ranked higher than a randomly chosen negative example.

**When to use AUC-PR instead?** On severely imbalanced datasets (e.g. fraud detection), the **Precision-Recall AUC** is more informative as it focuses on the minority class.

---

### Q8. What is the difference between Type I and Type II errors?

**Answer:**

| Error Type | Also Called | Definition | Example |
|:---|:---|:---|:---|
| **Type I Error** | False Positive (FP) | Rejecting a true null hypothesis | Spam filter marks a legitimate email as spam |
| **Type II Error** | False Negative (FN) | Failing to reject a false null hypothesis | Cancer test misses an actual cancer case |

- **Precision** is inversely related to Type I errors.
- **Recall** is inversely related to Type II errors.
- In high-stakes medical contexts, minimizing Type II errors (false negatives) is critical.

---

### Q9. What is log loss and why is it preferred over accuracy for probabilistic classifiers?

**Answer:**
**Log Loss (Binary Cross-Entropy)** penalizes confident wrong predictions more heavily than uncertain ones.

$$\text{Log Loss} = -\frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{p}_i) + (1 - y_i) \log(1 - \hat{p}_i) \right]$$

**Why prefer it over accuracy?**
- Accuracy only measures whether the final predicted label is correct.
- Log loss measures the **quality of the probability estimate**.
- Example: Predicting $p = 0.51$ vs $p = 0.99$ both give the same accuracy for a positive label, but log loss correctly penalizes the uncertain prediction.
- It rewards calibrated classifiers and is a smoother, differentiable objective for optimization.


## 3. Supervised Learning Algorithms

---

### Q10. How does a Random Forest work, and why does it reduce variance?

**Answer:**
A **Random Forest** is an ensemble of decision trees trained using **Bagging (Bootstrap Aggregating)**:

1. For each tree $t = 1 \ldots T$:
   - Sample a **bootstrap dataset** (random sample with replacement from training data).
   - At **each split**, only consider a random subset of features ($\sqrt{p}$ for classification, $p/3$ for regression where $p$ = total features).
   - Grow the full tree without pruning.
2. **Final prediction**: Majority vote (classification) or mean (regression) across all trees.

**Why it reduces variance:**
- Individual trees have low bias but high variance.
- The bootstrap sampling and random feature selection decorrelate the trees.
- Averaging decorrelated high-variance models reduces variance significantly without increasing bias.

$$\text{Var}(\bar{X}) = \frac{\sigma^2}{T} + \frac{T-1}{T} \cdot \rho \sigma^2$$
Where $\rho$ is the correlation between trees. Random feature selection reduces $\rho$.

---

### Q11. Explain gradient boosting and how it differs from bagging.

**Answer:**

| Property | Bagging (Random Forest) | Boosting (GBM, XGBoost) |
|:---|:---|:---|
| **Training** | Parallel (independent trees) | Sequential (each tree corrects previous) |
| **Focus** | Reduces variance | Reduces bias |
| **Data sampling** | Bootstrap samples | Weighted samples (or residuals) |
| **Base learner** | Full-depth trees (low bias) | Shallow trees / stumps (high bias) |
| **Overfitting** | Robust | Can overfit (needs early stopping) |

**Gradient Boosting mechanics:**
1. Initialize prediction: $F_0(x) = \bar{y}$
2. For each iteration $m$:
   - Compute **pseudo-residuals** (negative gradient of loss w.r.t. predictions): $r_i = -\frac{\partial L(y_i, F(x_i))}{\partial F(x_i)}$
   - Fit a tree $h_m$ to the residuals.
   - Update: $F_m(x) = F_{m-1}(x) + \eta \cdot h_m(x)$ (where $\eta$ is learning rate)

**XGBoost** adds L1/L2 regularization, second-order gradients, and efficient column block structures for speed.

---

### Q12. What is the kernel trick in SVMs?

**Answer:**
Support Vector Machines find the optimal hyperplane that maximizes the margin between classes. For non-linearly separable data, the **kernel trick** maps input features into a higher-dimensional space where a linear separator exists — without explicitly computing the transformation.

$$K(x_i, x_j) = \phi(x_i)^T \phi(x_j)$$

The SVM only needs dot products (through the kernel), not the full mapping $\phi$. Common kernels:

| Kernel | Formula | When to Use |
|:---|:---|:---|
| **Linear** | $x_i^T x_j$ | Linearly separable data |
| **Polynomial** | $(x_i^T x_j + c)^d$ | Polynomial boundaries |
| **RBF / Gaussian** | $\exp(-\gamma \|x_i - x_j\|^2)$ | General-purpose, most popular |
| **Sigmoid** | $\tanh(\kappa x_i^T x_j + c)$ | Neural-net-like boundary |

**Intuition**: Imagine points on a 1D line that can't be separated linearly — add a 2nd dimension $x^2$, and suddenly a separating hyperplane exists.

---

### Q13. Explain the difference between Generative and Discriminative models.

**Answer:**

| Property | Generative | Discriminative |
|:---|:---|:---|
| **What it models** | Joint distribution $P(X, Y)$ | Conditional distribution $P(Y \mid X)$ |
| **Can generate data** | ✅ Yes | ❌ No |
| **Examples** | Naive Bayes, HMM, VAE, GANs, LLMs | Logistic Regression, SVM, CRF, BERT |
| **Data efficiency** | Better with less data | Better with more data |
| **Handles missing features** | ✅ Naturally | ❌ Harder |

**Intuition:**
- Generative: "How was this data generated? Learn the full distribution."
- Discriminative: "I only care about boundaries between classes. Learn the decision boundary directly."


## 4. Unsupervised Learning

---

### Q14. How does K-Means clustering work and what are its limitations?

**Answer:**
**K-Means Algorithm:**
1. Initialize $k$ cluster centroids randomly.
2. **Assignment step**: Assign each point to the nearest centroid: $c_i = \arg\min_k \|x_i - \mu_k\|^2$
3. **Update step**: Recompute centroids as the mean of all assigned points: $\mu_k = \frac{1}{|C_k|} \sum_{i \in C_k} x_i$
4. Repeat until convergence (centroids don't move).

**Minimizes**: Within-cluster sum of squares (WCSS / inertia):
$$J = \sum_{k=1}^{K} \sum_{i \in C_k} \|x_i - \mu_k\|^2$$

**Limitations:**
- Must specify $k$ upfront (use **Elbow Method** or **Silhouette Score** to choose).
- Assumes **spherical clusters** of similar sizes.
- Sensitive to **outliers** and **random initialization** (use k-means++).
- Struggles with non-convex cluster shapes (use DBSCAN instead).
- Only works with **numerical** features.

---

### Q15. What is PCA and what is it used for?

**Answer:**
**Principal Component Analysis (PCA)** is a linear dimensionality reduction technique that projects data onto a lower-dimensional subspace while preserving maximum variance.

**Steps:**
1. Standardize the data.
2. Compute the covariance matrix $\Sigma = \frac{1}{N} X^T X$.
3. Compute eigenvectors (principal components) and eigenvalues of $\Sigma$.
4. Sort by descending eigenvalue (explained variance).
5. Project data onto top-$k$ eigenvectors.

**Use cases:**
- Dimensionality reduction before ML modeling.
- Noise reduction.
- Visualization (reducing to 2D/3D).
- Speeding up training.
- Removing multicollinearity.

**Key question in interviews**: "How many components to keep?"
→ Keep enough to explain 90-95% of the total variance (scree plot or cumulative explained variance curve).

**Limitations**: Linear only; use **t-SNE** or **UMAP** for non-linear dimensionality reduction / visualization.


## 5. Deep Learning & Neural Networks

---

### Q16. Explain Backpropagation intuitively.

**Answer:**
Backpropagation is the algorithm used to compute gradients of the loss function with respect to all weights in a neural network, enabling gradient descent parameter updates.

**Forward Pass**: Input flows through the network layer by layer to produce a prediction and loss.

**Backward Pass** (Backprop):
1. Compute the gradient of the loss w.r.t. the final layer output.
2. Using the **chain rule** of calculus, propagate gradients backwards through each layer:
$$\frac{\partial L}{\partial w_l} = \frac{\partial L}{\partial a_{l+1}} \cdot \frac{\partial a_{l+1}}{\partial z_l} \cdot \frac{\partial z_l}{\partial w_l}$$

3. Update weights: $w \leftarrow w - \eta \cdot \frac{\partial L}{\partial w}$

**Intuition**: Each weight's gradient measures "how much did this weight contribute to the final error?" Backprop efficiently computes this for every weight in a single backward pass through dynamic programming.

---

### Q17. What are vanishing and exploding gradients and how do you address them?

**Answer:**
When training deep networks, gradients flow back through many layers. Multiplying many small values (< 1) causes **vanishing gradients**; multiplying many large values (> 1) causes **exploding gradients**.

| Problem | Symptom | Solutions |
|:---|:---|:---|
| **Vanishing Gradients** | Early layers don't learn; training stalls | ReLU activations, Batch Norm, Residual connections (skip connections), LSTM/GRU for RNNs |
| **Exploding Gradients** | Loss becomes NaN; weights blow up | Gradient clipping, Weight initialization (Xavier/He), Batch Norm |

**Gradient Clipping**: If $\|\nabla\| > \text{threshold}$, scale: $\nabla \leftarrow \nabla \cdot \frac{\text{threshold}}{\|\nabla\|}$

---

### Q18. Explain Batch Normalization and why it helps training.

**Answer:**
**Batch Normalization** normalizes the inputs to each layer within a mini-batch, then applies learnable scale ($\gamma$) and shift ($\beta$) parameters.

$$\hat{x}_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}, \quad y_i = \gamma \hat{x}_i + \beta$$

**Benefits:**
1. **Stabilizes training**: Reduces internal covariate shift (distribution of layer inputs shifting during training).
2. **Allows higher learning rates**: Gradients are better scaled.
3. **Regularization effect**: Acts as a weak regularizer (reduces need for dropout).
4. **Reduces sensitivity to initialization**.

**Inference**: Uses running mean/variance (computed during training), not batch statistics.

---

### Q19. What is the attention mechanism and why is it important?

**Answer:**
**Attention** allows a model to selectively focus on relevant parts of the input when producing each output, instead of compressing all information into a fixed-size vector (the bottleneck of seq2seq RNNs).

**Scaled Dot-Product Attention:**
$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right) V$$

Where:
- $Q$ (Queries): What we're looking for.
- $K$ (Keys): What each element offers.
- $V$ (Values): The actual content.
- $\sqrt{d_k}$: Scaling factor to prevent softmax saturation.

**Multi-Head Attention** runs $h$ attention heads in parallel, letting the model attend to different representation subspaces simultaneously.

**Why it transformed NLP**: Transformers (BERT, GPT) replaced RNNs entirely with attention, enabling:
- **Parallelization** (no sequential dependency).
- Capturing **long-range dependencies**.
- Better scalability to large datasets.

---

### Q20. What is the difference between BERT and GPT? When would you use each?

**Answer:**

| Property | BERT | GPT |
|:---|:---|:---|
| **Architecture** | Encoder-only Transformer | Decoder-only Transformer |
| **Training Objective** | Masked Language Model (MLM) + NSP | Causal Language Modeling (CLM) |
| **Attention** | Bidirectional (sees full context) | Unidirectional (left-to-right only) |
| **Strength** | Understanding / Encoding text | Generating text |
| **Fine-tune use cases** | Classification, NER, Q&A, Sentence Similarity | Text generation, summarization, chat, code |
| **Examples** | BERT, RoBERTa, DeBERTa, ALBERT | GPT-2, GPT-4, LLaMA, Mistral |

**BERT Pretraining (MLM)**: Randomly mask 15% of tokens and predict them: `[CLS] The cat [MASK] on the mat [SEP]` → predict "sat".

**GPT Pretraining (CLM)**: Predict the next token given all previous ones: $P(w_t | w_1, \ldots, w_{t-1})$.


## 6. Natural Language Processing (NLP)

---

### Q21. What is word embedding and how is Word2Vec trained?

**Answer:**
**Word Embeddings** are dense, low-dimensional vector representations of words that capture semantic relationships.

**Word2Vec** trains embeddings using a shallow 2-layer neural network. Two architectures:

1. **CBOW (Continuous Bag of Words)**: Predict the target word from surrounding context words.
2. **Skip-gram**: Predict surrounding context words from a target word. (Better for rare words)

**Training trick — Negative Sampling**: Instead of computing softmax over entire vocabulary (slow), train the model to distinguish real (positive) context pairs from randomly sampled (negative) noise pairs.

**Resulting properties:**
- Semantic relationships encoded as vector arithmetic: `king - man + woman ≈ queen`
- Similar words cluster in the embedding space.

**Limitation**: Context-free (one vector per word regardless of meaning). → Replaced by **contextual embeddings** (ELMo, BERT).

---

### Q22. What is fine-tuning a pretrained language model? What are PEFT methods?

**Answer:**
**Fine-tuning**: Take a pretrained model (e.g., BERT, LLaMA) and continue training on a smaller, task-specific dataset, updating all or some weights to specialize the model.

**Full fine-tuning**: All parameters are updated. Expensive in compute and storage for large models.

**PEFT (Parameter-Efficient Fine-Tuning)**: Only update a small subset of parameters.

| Method | Description | Parameters Trained |
|:---|:---|:---|
| **LoRA** | Injects low-rank decomposition matrices $W = W_0 + AB$ into attention layers | $A$ and $B$ only (~0.1-1% of model) |
| **Prefix Tuning** | Prepend learnable "soft prompt" tokens to input | Prefix tokens only |
| **Prompt Tuning** | Learnable prompt embeddings at input layer | Even fewer parameters |
| **Adapter Layers** | Insert small bottleneck layers inside transformer blocks | Adapter weights only |

**LoRA** is the dominant PEFT method today, used in tools like Hugging Face PEFT and Axolotl.

---

### Q23. What is RAG (Retrieval-Augmented Generation)?

**Answer:**
**RAG** combines a retrieval system with a language model to answer questions using external knowledge without retraining.

**Pipeline:**
1. **Indexing**: Chunk documents → embed with an encoder (e.g., `sentence-transformers`) → store in a vector database (Faiss, Pinecone, Chroma).
2. **Query time**:
   - Embed the user query.
   - Retrieve top-$k$ most similar document chunks (approximate nearest-neighbor search).
   - Feed retrieved context + query as a prompt to an LLM.
   - LLM generates a grounded answer.

**Why RAG?**
- Overcomes LLM **knowledge cutoff** (access to real-time/private data).
- Reduces **hallucinations** by grounding generation in retrieved facts.
- More **cost-efficient** than full fine-tuning for knowledge injection.

**Challenges**: Retrieval quality (chunking strategy, embedding model choice), context window limits, faithfulness vs. fluency tradeoff.


## 7. MLOps & Production ML

---

### Q24. What is MLOps? What problem does it solve?

**Answer:**
**MLOps** (Machine Learning Operations) is a set of practices that aims to deploy and maintain ML models in production reliably and efficiently, combining ML engineering with DevOps principles.

**The core problem**: The gap between "model works in a notebook" and "model reliably serves millions of requests in production".

**MLOps covers:**

| Area | Key Concerns |
|:---|:---|
| **Data Management** | Versioning, lineage, validation, pipelines |
| **Experiment Tracking** | Logging metrics, hyperparameters, artifacts (MLflow, W&B) |
| **Model Registry** | Versioning, staging, promotion |
| **CI/CD for ML** | Automated retraining, testing, deployment pipelines |
| **Model Serving** | REST APIs, latency, throughput, batching (TorchServe, TF Serving, vLLM) |
| **Monitoring** | Data drift, model drift, performance degradation |
| **Governance** | Fairness, explainability, audit trails |

---

### Q25. What is model drift and how do you detect and handle it?

**Answer:**
**Model drift** refers to degradation of model performance over time because the real-world data distribution shifts from the training distribution.

**Types:**

| Type | Definition | Example |
|:---|:---|:---|
| **Data drift** (Covariate shift) | Input feature distribution $P(X)$ changes | User demographics shift over time |
| **Concept drift** | Relationship $P(Y \| X)$ changes | Consumer behavior changes post-pandemic |
| **Label drift** | Output distribution $P(Y)$ changes | Fraud patterns evolve |
| **Upstream data drift** | Schema / pipeline changes alter raw data | API returns new fields / null values |

**Detection:**
- Statistical tests: **KS test**, **Chi-squared test**, **PSI (Population Stability Index)**, **Jensen-Shannon divergence**.
- Monitor feature statistics (mean, std, null rates, quantiles) vs. training baseline.
- Track model performance metrics (if labels are available with delay).

**Handling:**
- **Retraining triggers**: Scheduled retraining, performance-threshold-based, or drift-score-based.
- **Online learning**: Continuously update model with new data.
- **Champion/Challenger** A/B testing when deploying new model versions.

---

### Q26. Explain the difference between batch inference and real-time (online) inference.

**Answer:**

| Property | Batch Inference | Real-Time / Online Inference |
|:---|:---|:---|
| **Trigger** | Scheduled (e.g., nightly job) | On-demand (per request) |
| **Latency requirement** | Minutes to hours | Milliseconds to hundreds of ms |
| **Throughput** | Very high (process millions at once) | Lower per-instance throughput |
| **Infrastructure** | Spark, Ray, Dataflow, batch ML pipelines | REST APIs, gRPC, Kafka + streaming |
| **Cost** | Cheaper (spot instances, batch pricing) | Higher (always-on servers) |
| **Use cases** | Recommendations (next-day batch), fraud scoring of past transactions | Search ranking, content moderation, chatbots |

**Hybrid — Near-Real-Time**: Pre-compute predictions for common cases (cache/warm), compute on-demand for tail cases.

---

### Q27. What is a Feature Store and why is it important in MLOps?

**Answer:**
A **Feature Store** is a centralized repository for storing, serving, and sharing ML features.

**Problem it solves:**
- Teams recompute the same features multiple times → waste.
- Training features and serving features diverge → **training-serving skew**.
- Features aren't reusable across teams and models.

**Components:**

| Layer | Description |
|:---|:---|
| **Offline store** | Historical feature data (Parquet/Delta Lake) for training |
| **Online store** | Low-latency key-value store (Redis, DynamoDB) for serving |
| **Feature pipeline** | Transforms raw data into features (Spark, Flink, dbt) |
| **Feature registry** | Metadata, lineage, ownership documentation |

**Examples**: Feast, Tecton, Hopsworks, Vertex AI Feature Store.

**Critical concept**: **Point-in-time correctness** — when creating training data, features must only use information available *at the time of the label event* to prevent data leakage.

---

### Q28. How do you approach deploying an ML model with CI/CD?

**Answer:**
A mature ML CI/CD pipeline includes:

```
Code Push
    │
    ▼
CI Pipeline
├── Unit tests (data validation, feature transforms, model code)
├── Integration tests (pipeline end-to-end)
├── Training job (on subset or full data)
├── Model evaluation (metrics vs. baseline / previous champion)
│
    ▼ (if metrics pass)
CD Pipeline
├── Package model artifact (Docker image / SavedModel)
├── Push to Model Registry (tagged with version, metrics, data hash)
├── Deploy to staging environment
├── Shadow mode / A-B test
├── Canary rollout (5% → 25% → 100%)
└── Monitor & rollback if drift detected
```

**Key MLOps practices:**
- **Model versioning**: Every model has a reproducible lineage (data version + code version + hyperparameters).
- **Reproducibility**: Fix random seeds, pin dependency versions.
- **Rollback**: Keep previous model version deployed; switch back instantly on regression.


## 8. Feature Engineering & Data Preprocessing

---

### Q29. How do you handle missing data?

**Answer:**

| Strategy | When to Use | Notes |
|:---|:---|:---|
| **Drop rows** | Missing completely at random (MCAR), very few rows | Risk of data loss |
| **Drop columns** | > 50-70% of values missing | If feature adds little value |
| **Mean/Median imputation** | MCAR, numerical features | Median for skewed distributions |
| **Mode imputation** | Categorical features | |
| **KNN imputation** | Missing at random (MAR), smaller datasets | Expensive to compute |
| **Model-based imputation** | MAR, complex patterns | Use another ML model to predict missing values |
| **Add missingness indicator** | Missingness carries signal | Add binary `feature_is_null` column alongside |
| **Multiple Imputation** | Formal statistical analysis | Creates multiple imputed datasets |

**Key insight**: Always analyze *why* data is missing (MCAR/MAR/MNAR) before choosing a strategy, and avoid introducing data leakage during imputation (fit imputer only on training data).

---

### Q30. How do you handle class imbalance?

**Answer:**

| Strategy | Description |
|:---|:---|
| **Oversampling minority** | SMOTE, ADASYN — generate synthetic minority samples |
| **Undersampling majority** | Random undersampling, Tomek links |
| **Class weights** | Most sklearn/PyTorch models support `class_weight='balanced'` |
| **Threshold tuning** | Adjust decision threshold instead of 0.5 (optimize F1, recall) |
| **Ensemble methods** | BalancedBaggingClassifier, EasyEnsemble |
| **Anomaly detection framing** | Treat minority as outliers (if ratio < 1:100) |

**Always**: Use stratified splits and report Precision-Recall AUC, not accuracy.

---

### Q31. What is target encoding and when should you use it?

**Answer:**
**Target Encoding** replaces each category with the mean of the target variable for that category:

$$\text{encoded}(c) = \frac{\sum_{i: x_i=c} y_i}{|\{i: x_i = c\}|}$$

**Advantages:**
- Handles high-cardinality categoricals (e.g., ZIP codes, user IDs) where one-hot encoding explodes.
- Captures the relationship between category and target.

**Danger — Data Leakage**: If done naively on the full training set, the encoding carries target signal. **Use cross-fold target encoding** (compute encoding using all folds except the current one, similar to stacking).

**Smoothing** (to handle rare categories):
$$\text{smoothed}(c) = \frac{n_c \cdot \mu_c + m \cdot \mu_{\text{global}}}{n_c + m}$$
Where $m$ is a smoothing factor. Rare categories are pulled toward the global mean.

### Q32. How would you design a recommendation system?

**Answer:**
A production recommendation system typically uses a multi-stage pipeline:

```
User Request
     │
     ▼
1. CANDIDATE GENERATION (Retrieval)
   ├── Collaborative Filtering (Matrix Factorization, ALS)
   ├── Content-Based (item embeddings, TF-IDF similarity)
   └── Two-Tower Neural Model → approximate NN search (FAISS)
   (Retrieves ~100-1000 candidates efficiently from millions of items)
     │
     ▼
2. RANKING (Scoring)
   ├── Gradient Boosted Tree (XGBoost, LightGBM) or
   └── Deep neural ranker (Wide & Deep, DCN)
   Features: user history, item metadata, context (time, device)
   (Re-ranks the ~100-1000 candidates with a more expensive model)
     │
     ▼
3. BUSINESS LOGIC / FILTERS
   ├── Diversity / deduplication
   ├── Freshness boost
   └── Compliance / fairness filters
     │
     ▼
4. SERVING
   └── Return top-K recommendations
```

**Key challenges**: Cold start (new users/items), feedback loops (popularity bias), online evaluation (A/B tests, interleaving).

### Q33. How would you design a real-time fraud detection system?

**Answer:**

**Requirements:**
- Ultra-low latency (< 100ms per transaction).
- High precision AND recall (miss fraud = loss; flag legitimate = bad UX).
- Handle 10K-100K transactions/second.
- Handle class imbalance (< 1% fraud).

**Architecture:**

```
Transaction Event
       │
       ▼
Feature Engineering Layer
├── Real-time features: transaction amount, merchant, location, time
├── User historical features: avg spend, usual location (Feature Store - Redis)
└── Entity features: merchant risk score, device fingerprint
       │
       ▼
Scoring Layer
├── Fast Rule Engine (threshold checks, velocity rules)  → immediate block for obvious fraud
└── ML Model (LightGBM or small neural net) → probability score
       │
       ▼
Decision Layer
├── Score < 0.3 → Approve
├── 0.3 – 0.7 → Step-up authentication (2FA)
└── Score > 0.7 → Decline / Flag for review
       │
       ▼
Post-Transaction (async)
├── Log to feature store
├── Analyst review queue
└── Label feedback → Retraining pipeline
```

**Key design choices:**
- Use **online feature store** (Redis) for sub-10ms feature lookup.
- **Champion/Challenger** A/B testing for new model rollouts.
- **Concept drift monitoring**: Fraud patterns evolve rapidly; monitor weekly.
