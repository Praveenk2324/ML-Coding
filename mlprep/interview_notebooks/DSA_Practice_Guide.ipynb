## Module 1: Arrays & Hashing

Arrays and Hash tables are the bedrock of DSA. Using a hash map (dictionary in Python) allows you to achieve $O(1)$ lookups, which is a powerful way to reduce the time complexity of search operations from $O(N^2)$ to $O(N)$.

---

### Solved Demo: Two Sum
**Problem Statement**: Given an array of integers `nums` and an integer `target`, return indices of the two numbers such that they add up to `target`.
* **Complexity**: Time $O(N)$ | Space $O(N)$

```python
def twoSum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

# Test the demo
print("Two Sum Demo Output (expected [0, 1]):", twoSum([2, 7, 11, 15], 9))
```

## Module 2: Two Pointers

The two-pointer technique uses two indices to traverse a linear data structure (like an array or string) simultaneously. This is often used on sorted arrays to search for pairs or to process elements in $O(N)$ time with $O(1)$ space.

---

### Solved Demo: Valid Palindrome
**Problem Statement**: Determine if a string `s` is a palindrome, considering only alphanumeric characters and ignoring cases.
* **Complexity**: Time $O(N)$ | Space $O(1)$

```python
def isPalindrome(s):
    l, r = 0, len(s) - 1
    while l < r:
        while l < r and not s[l].isalnum():
            l += 1
        while l < r and not s[r].isalnum():
            r -= 1
        if s[l].lower() != s[r].lower():
            return False
        l += 1
        r -= 1
    return True

# Test the demo
print("Palindrome Demo Output (expected True):", isPalindrome("A man, a plan, a canal: Panama"))
```

## Module 3: Sliding Window

Sliding Window is a sub-category of the Two Pointers technique where the distance between the two pointers (window size) expands or contracts based on constraints. It is highly effective for subarray or substring search problems to avoid nested loops.

---

### Solved Demo: Best Time to Buy and Sell Stock
**Problem Statement**: Find the maximum profit you can achieve by buying and selling a single stock on different days.
* **Complexity**: Time $O(N)$ | Space $O(1)$

```python
def maxProfit(prices):
    min_price = float('inf')
    max_profit = 0
    for price in prices:
        if price < min_price:
            min_price = price
        elif price - min_price > max_profit:
            max_profit = price - min_price
    return max_profit

# Test the demo
print("Best Stock Profit Demo Output (expected 5):", maxProfit([7, 1, 5, 3, 6, 4]))
```

A Linked List is a linear data structure where elements are not stored at contiguous memory locations. Linked list nodes contain data and a pointer to the next node. Common operations include node re-linking and using slow/fast pointer strategies.

### Singly Linked List Node Definition
We define a helper class `ListNode` to represent nodes and provide utilities to convert lists to and from Python arrays.

---

### Solved Demo: Reverse Linked List
**Problem Statement**: Reverse a singly linked list in-place and return the new head.
* **Complexity**: Time $O(N)$ | Space $O(1)$

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    # Utility to convert node chain to a python list
    def to_list(self):
        res = []
        curr = self
        while curr:
            res.append(curr.val)
            curr = curr.next
        return res

    @staticmethod
    def from_list(lst):
        if not lst:
            return None
        head = ListNode(lst[0])
        curr = head
        for val in lst[1:]:
            curr.next = ListNode(val)
            curr = curr.next
        return head

def reverseList(head):
    prev = None
    curr = head
    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt
    return prev

# Test the demo
head = ListNode.from_list([1, 2, 3, 4, 5])
reversed_head = reverseList(head)
print("Reversed List Demo Output (expected [5, 4, 3, 2, 1]):", reversed_head.to_list())
```

## Module 5: Trees & Graphs

Binary Trees are hierarchical structures where each node has at most two children. Tree traversal algorithms (DFS, BFS) are fundamental.

---

### Tree Node Definition
We define a helper class `TreeNode` representing nodes and utilities to convert to and from level-order lists.

---

### Solved Demo: Invert Binary Tree
**Problem Statement**: Invert a binary tree (swap left and right child recursively) and return its root.
* **Complexity**: Time $O(N)$ | Space $O(H)$ where $H$ is tree height.

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def to_level_order(self):
        res = []
        queue = [self]
        while queue:
            node = queue.pop(0)
            if node:
                res.append(node.val)
                queue.append(node.left)
                queue.append(node.right)
            else:
                res.append(None)
        while res and res[-1] is None:
            res.pop()
        return res

    @staticmethod
    def from_level_order(lst):
        if not lst:
            return None
        root = TreeNode(lst[0])
        queue = [root]
        i = 1
        while queue and i < len(lst):
            node = queue.pop(0)
            if lst[i] is not None:
                node.left = TreeNode(lst[i])
                queue.append(node.left)
            i += 1
            if i < len(lst) and lst[i] is not None:
                node.right = TreeNode(lst[i])
                queue.append(node.right)
            i += 1
        return root

def invertTree(root):
    if not root:
        return None
    root.left, root.right = invertTree(root.right), invertTree(root.left)
    return root

# Test the demo
tree = TreeNode.from_level_order([4, 2, 7, 1, 3, 6, 9])
inverted = invertTree(tree)
print("Inverted Tree Demo Output (expected [4, 7, 2, 9, 6, 3, 1]):", inverted.to_level_order())
```

## Module 6: Dynamic Programming

Dynamic Programming (DP) solves complex problems by breaking them down into simpler subproblems. It solves each subproblem once and stores the result (memoization/tabulation) to avoid redundant calculations.

---

### Solved Demo: Climbing Stairs
**Problem Statement**: You are climbing a staircase. It takes `n` steps to reach the top. Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?
* **Complexity**: Time $O(N)$ | Space $O(1)$

```python
def climbStairs(n):
    if n <= 2:
        return n
    one, two = 1, 2
    for _ in range(3, n + 1):
        one, two = two, one + two
    return two

# Test the demo
print("Climbing 5 Stairs Demo Output (expected 8):", climbStairs(5))
```
